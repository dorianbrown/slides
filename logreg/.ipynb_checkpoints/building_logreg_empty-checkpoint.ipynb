```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

np.random.seed(9001)

data = np.array(pd.read_csv("classification.csv"))
X = data[:,0:2]
y = data[:,2]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
class LogisticRegression:
    def __init__(self, lr=0.01, num_iter=100000):
        self.lr = lr
        self.num_iter = num_iter
    
    def __sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def __loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    def fit(self, X, y):
        # Adding intercept term to X
        X = np.concatenate((X, np.ones((X.shape[0],1))), axis = 1)
        # weights initialization
        self.theta = np.zeros(X.shape[1])
        
        for i in range(self.num_iter):
            z = np.dot(X, self.theta)
            h = self.__sigmoid(z)
            gradient = np.dot(X.T, (h - y)) / y.size
            self.theta -= self.lr * gradient
            
            if i % 10000 == 0:
                z = np.dot(X, self.theta)
                h = self.__sigmoid(z)
                print(f'loss: {self.__loss(h, y)} \t')
                print(f'theta: {self.theta}')
    
    def predict_prob(self, X):
        X = np.concatenate((X, np.ones((X.shape[0],1))), axis = 1)
        return self.__sigmoid(np.dot(X, self.theta))
    
    def predict(self, X, threshold):
        return self.predict_prob(X) >= threshold
```

```python
lr = LogisticRegression(lr=0.1)

lr.fit(X_train, y_train)

pred_y = lr.predict(X_test, threshold=0.5)

plt.scatter(x = X_test[:,0], y = X_test[:,1], c = pred_y)
```

```python
X_nonlinear = np.hstack((X, X[:,0:1]**2, X[:,1:2]**2, X[:,1:2]*X[:,0:1], X[:,0:1]**3, X[:,1:2]**3, np.log(X[:,0:1]), np.log(X[:,1:2])))

from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(3)

X_overfit = pf.fit_transform(X_nonlinear)

X_train, X_test, y_train, y_test = train_test_split(X_overfit, y, test_size=0.25)
```
